```
import pandas as pd
import matplotlib.pyplot as plt
```

***IRIS DATASET***

```
df=pd.read_csv("iris.csv")
```

```
df.head()
```

```
 sepal_width=df.loc[:,["sepal_width"]]
  sepal_length=df.loc[:,["sepal_length"]]
```

***histogram***

```
plt.hist(sepal_width,density=True, facecolor='g', alpha=1)
plt.xlabel("sepal width")
plt.ylabel("frequency")
plt.title("iris dataset")
plt.show()
```

***step plot***

```
plt.step(sepal_length,sepal_width)
plt.plot(sepal_length, sepal_width, 'C1o', alpha=0.5)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("iris dataset")
plt.show
```

***scatter plot***

```
plt.scatter(sepal_length, sepal_width, marker='^',alpha=0.5)
plt.grid(True)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.title("iris data")
plt.show()
```

***violin plot***

```
import seaborn as sns
sns.violinplot(sepal_length,color="yellow")
```

```
plt.plot(sepal_length,sepal_width)
plt.show()
```
